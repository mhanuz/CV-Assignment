```python
#import library
import cv2 as cv
from matplotlib import pyplot as plt

#upload image 
img = cv.imread('images/berry-1.jpg', cv.IMREAD_GRAYSCALE)

# adaptive threshoding method 
thresh1 = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY,85,15)
thresh2 = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY,95,15)

# plotting 
plt.figure(figsize=(8, 10))

plt.subplot(1, 2, 1)
plt.imshow(thresh1, 'gray')
plt.title('Adaptive mean Thresholding')

plt.subplot(1, 2, 2)
plt.imshow(thresh2, 'gray')
plt.title('Adaptive Gaussian Thresholding')
```
