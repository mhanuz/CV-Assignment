## OpenCV

```python
import cv2
image_path = 'images/sunflower.jfif'
image = cv2.imread(image_path)
cv2.imshow('sunflower', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Skimage

```python
import skimage
from skimage import io

img = io.imread('images/sunflower.jfif')
io.imshow(img)
io.show()
```
