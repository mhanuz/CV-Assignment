## SkImage

```python
from skimage import io
from matplotlib import pyplot as plt
```

```python
rgb_image = io.imread('images/sunflower.jfif')
```

```python
rgb_image.shape
```

```python
red_image = rgb_image[:,:,0]
green_image = rgb_image[:,:,1] 
blue_image = rgb_image[:,:,2]
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(2,2,1)
plt.imshow(rgb_image)
plt.title("Original Image")

plt.subplot(2,2,2)
plt.imshow(red_image, 'gray')
plt.title("Red Channel")

plt.subplot(2,2,3)
plt.imshow(green_image, 'gray')
plt.title("Green Channel")

plt.subplot(2,2,4)
plt.imshow(blue_image, 'gray')
plt.title("Blue Channel")
```

## Pillow

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = Image.open("images/sunflower.jfif")
```

```python
image_array = np.asarray(img)
```

```python
red_image = image_array[:,:,0]
green_image = image_array[:,:,1]
blue_image = image_array[:,:,2]
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(2,2,1)
plt.imshow(img)
plt.title("Original Image")

plt.subplot(2,2, 2)
plt.imshow(red_image, cmap='gray')
plt.title("Red Channel")

plt.subplot(2,2,3)
plt.imshow(green_image, cmap='gray')
plt.title("Green Channel")

plt.subplot(2,2,4)
plt.imshow(blue_image, cmap='gray')
plt.title("Blue Channel")

plt.show()
```
