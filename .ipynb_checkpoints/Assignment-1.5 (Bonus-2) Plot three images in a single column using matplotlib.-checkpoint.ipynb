```python
import matplotlib.pyplot as plt
```

```python
Mash_image = plt.imread('images/mashrafee.jfif')
Shakib_image = plt.imread('images/shakib.jfif')
Tamim_image = plt.imread('images/tamim.jfif')
```

```python
plt.figure(figsize=(10,10))

plt.subplot(1,3,1)
plt.imshow(Mash_image)
plt.title("Mashrafee")

plt.subplot(1,3,2)
plt.imshow(Shakib_image)
plt.title("Shakib_image")

plt.subplot(1,3,3)
plt.imshow(Tamim_image)
plt.title("Tamim_image")
```
